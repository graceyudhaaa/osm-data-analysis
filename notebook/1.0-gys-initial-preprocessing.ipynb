```python
import pandas as pd
import numpy as np
```

## Loading The Dataset

```python
df = pd.read_excel("../data/raw/OSM Data.xlsx")
df.head()
```

## Data Cleaning

```python
df.columns
```

### renaming column to snake_case

```python
column=['home', 'away', 'stronger_team', 'your_team', 'home_formation',
        'away_formation', 'home_style_of_play', 'home_marking',
        'home_offside_trap', 'home_tackling', 'away_style_of_play',
        'away_marking', 'away_offside_trap', 'away_tackling',
        'home_training_camp', 'away_training_camp', 'line_tactic_FW',
        'line_tactic_MF', 'line_tactic_DF', 'pressing', 'style', 'tempo',
        'home_posession', 'away_posession', 'home_goals', 'away_goals',
        'home_shots', 'away_shots', 'tip', 'result', 'your_result']

df.columns = column
df.head()
```

```python
df.to_csv("../data/raw/OSM Data.csv", index=False, header=True)
```

### casefolding the string to lowercase

```python
for col in df.columns:
    if col == 'tip':
        continue
    if df[col].dtypes == 'O':
        df[col] = df[col].apply(lambda x: x.lower())

df.head()
```

```python
# drop unnecessary columns
df = df.drop(['home', 'away', 'tip'], axis=1)
df.head()
```

### map the team columns into opponent based columns instead home and away based

```python
# ternary in python: [on_true] if [expression] else [on_false] 
df['strength'] = df.apply(lambda row: 'equal' if (row['stronger_team'] == 'equal')\
                          else 'strong'\
                        if (row['stronger_team'] == row['your_team'])\
                            else 'weak', axis=1)
df = df.drop('stronger_team', axis=1)
df.head()
```

```python
home_away_col = [c for c in column if (len(c) > 4 and (c.startswith('home') or c.startswith('away')))]

home_away_prop = set([prop[5:] for prop in home_away_col])
home_away_prop
```

```python
def home_away_mapping(row, your_team, true_state_col, false_state_col):
    if (row['your_team'] == your_team):
        return row[true_state_col]
    else:
        return row[false_state_col]
```

```python
for prop in home_away_prop:
    df[prop] = df.apply(lambda row: home_away_mapping(row, 'home', f'home_{prop}', f'away_{prop}'), axis=1)
    df[f'opponent_{prop}'] = df.apply(lambda row: home_away_mapping(row, 'away', f'home_{prop}', f'away_{prop}'), axis=1)

df = df.drop(home_away_col, axis=1)
df.head()
```

### remove subjectiveness from result column

```python
df['result'] = df.apply(lambda row: row['your_result'], axis=1).map({'you won': 'won',
                                                                    'you lost': 'lost',
                                                                    'draw': 'draw'})
df = df.drop('your_result', axis=1)
df.head()
```

```python
# calculate goal different
df['goal_diff'] = np.abs(df['goals'] - df['opponent_goals'])
df.head()
```

```python
column_order = [
                    'strength',
                    'opponent_formation', 'opponent_style_of_play', 'opponent_marking', 'opponent_offside_trap', 'opponent_tackling', 'opponent_training_camp',
                    'formation', 'style_of_play', 'marking', 'offside_trap', 'tackling', 'training_camp',
                    'line_tactic_FW', 'line_tactic_MF', 'line_tactic_DF',
                    'pressing', 'style', 'tempo',
                    'goals', 'opponent_goals', 'goal_diff',
                    'posession', 'opponent_posession',
                    'shots', 'opponent_shots',
                    'result',
                ]  

df = df[column_order]
df.head()
```
